```python
for i in url:
    request=r.get(i)
    print(request)
```

```python
html=[]
for i in url:
    try:
        html.append(requests.get(i).content)
    except:
        html.append(0)
```

```python
player_name=[]
player_id=[]


for i in html:
    soup = (BeautifulSoup(i, "lxml"))
    player_names=soup.select("a.nowrap[title]")
    player_name.append([i.get('title') for i in player_names])
    player_ids=soup.select("tr>td.col-pi")
    player_id.append([i.text for i in player_ids])
```

```python
id=[]
name=[]
#[test.append(player_id[i][j]) for i in range(len(player_id)) for j in range(len(player_id[i]))]

for i in player_id:
    for j in i:
        id.append(j)
        
for i in player_name:
    for j in i:
        name.append(j)
```

```python
import pandas as pd
df1 = pd.DataFrame(list(zip(id,name)),columns =['id', 'name']) 
df1 
```
