# Web scraping - Producthunt.com

```python
import requests
from bs4 import BeautifulSoup
import urllib.request
import time
import requests as r
```

## Topic selection

```python
topics=['productivity','tech','artificial-intelligence','marketing','analytics','apis','education']
```

## Getting Data from topic pages

```python
url=[]
for i in topics:
    try:
        url.append('https://www.producthunt.com/topics/'+i)
    except:
        url.append(0)
        
url
```

```python
html=[]
for i in url:
    try:
        html.append(requests.get(i).content)
    except:
        html.append(0)

#html
```

```python
topic_name=[]
topic_desc=[]
products=[]
all_links = []

for i in html:
    soup = (BeautifulSoup(i, "lxml"))
    topic_name.append(soup.find('h1').text.strip())
    topic_desc.append(soup.find('h2').text.strip())
    links = soup.select("a.link_523b9[href]")
    all_links.append([i.get('href') for i in links])
    products.append([i.text for i in soup.find_all('h3')])
       
#products
```

```python
for i in range(len(all_links)):
    for j in range(len(all_links[i])):
        if all_links[i][j]=='/r/p/151401?ref=topic_page&promo_id=6859':
            all_links[i][j]='/posts/monday'
```

## Getting Data from Product pages

```python
url2 = []

for i in range(len(all_links)):
    for j in range(len(all_links[i])):
        url2.append(all_links[i][j])
```

```python
url2
```

```python
#Removing non compatible URL
url2.remove('https://www.producthunt.com/r/p/29143?ref=topic_page&promo_id=8245')
```

```python
import time
```

```python
html2=[]

for i in url2:
    try:
        request=r.get(i)
        print(request)
        html2.append(request.content)
        #if i%2==0:
            #time.sleep(2)
    except:
        html2.append(0)
```

```python
product_name=[]
product_desc=[]
product_links=[]
votes=[]
reviews=[]

#for ind,i in enumerate(html2):
    #if ind%2==0:
        
for i in html2:
    soup2 = (BeautifulSoup(i, "lxml"))
    product_name.append((soup2.find('h1')).text.strip())
    product_desc.append((soup2.find('h2')).text.strip())

    product_link = soup2.select("a[href^='/r/']")
    product_links.append([i.get('href') for i in product_link][0])

    number_votes = soup2.select("span[class^='bigButtonCount']")
    number_votes=[element.text for element in number_votes]
    votes.append(int(number_votes[0].replace(',','')))
    
    review = soup2.select("div[class^='reviewContainer']>div>span")
    review=[element.text for element in review]
    if len(review)==3:
        reviews.append(review[2]) 
    else:
        reviews.append('NaN') 
```

```python
for i in range(len(product_links)):
    product_links[i]=('https://www.producthunt.com')+product_links[i]
```
